```python
import numpy as np
import matplotlib.pyplot as plt

# Time range
t = np.linspace(0, 10, 100)

# Constants
k = 0.5               # Growth rate
alpha = 0.8           # Proportion given
C0_full = 5           # Starting with energy
C0_empty = 0          # Starting empty

# Energy over time
C_full = C0_full * np.exp(k * t)
C_empty = C0_empty * np.exp(k * t)

# Giving over time
G_full = alpha * C_full
G_empty = alpha * C_empty

# Derivatives: how fast energy and giving change
dC_full = C0_full * k * np.exp(k * t)
dC_empty = C0_empty * k * np.exp(k * t)
dG_full = alpha * dC_full
dG_empty = alpha * dC_empty

# Plotting
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot C(t)
axs[0, 0].plot(t, C_full, label="C = 5", color='blue')
axs[0, 0].plot(t, C_empty, label="C = 0", linestyle='--', color='red')
axs[0, 0].set_title("Energy Growth: C(t)")
axs[0, 0].set_xlabel("Time")
axs[0, 0].set_ylabel("C(t)")
axs[0, 0].legend()

# Plot G(t)
axs[0, 1].plot(t, G_full, label="G(t) from full cup", color='green')
axs[0, 1].plot(t, G_empty, label="G(t) from empty cup", linestyle='--', color='orange')
axs[0, 1].set_title("Giving Over Time: G(t)")
axs[0, 1].set_xlabel("Time")
axs[0, 1].set_ylabel("G(t)")
axs[0, 1].legend()

# Plot dC/dt
axs[1, 0].plot(t, dC_full, label="dC/dt (C=5)", color='purple')
axs[1, 0].plot(t, dC_empty, label="dC/dt (C=0)", linestyle='--', color='gray')
axs[1, 0].set_title("Rate of Energy Growth: dC/dt")
axs[1, 0].set_xlabel("Time")
axs[1, 0].set_ylabel("dC/dt")
axs[1, 0].legend()

# Plot dG/dt
axs[1, 1].plot(t, dG_full, label="dG/dt (C=5)", color='brown')
axs[1, 1].plot(t, dG_empty, label="dG/dt (C=0)", linestyle='--', color='pink')
axs[1, 1].set_title("Rate of Giving: dG/dt")
axs[1, 1].set_xlabel("Time")
axs[1, 1].set_ylabel("dG/dt")
axs[1, 1].legend()

plt.tight_layout()
plt.show
```
